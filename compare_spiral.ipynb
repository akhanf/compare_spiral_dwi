```python
import pandas as pd

csv='results/{method}/sub-P156/dwi/sub-P156_atlas-schaefer_struc.conn.csv'
methods = ['spiral','epi']

dfs={}
for method in methods:
    dfs[method] = pd.read_csv(csv.format(method=method))
    
        
```

```python
import matplotlib.pyplot as plt
import numpy as np

conn_epi = dfs['epi'].to_numpy()
conn_spiral = dfs['spiral'].to_numpy()
```

```python
import matplotlib as mpl

cmap = mpl.cm.get_cmap("viridis")
cmap.set_under(color='grey')    

imshow_opts={'vmin':0.0000001,'vmax':500,'interpolation': 'none','cmap':cmap}



fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,8))


im = axes[0].imshow(conn_epi,**imshow_opts)
im = axes[1].imshow(conn_spiral,**imshow_opts)
axes[0].set_title('non-spiral')
axes[1].set_title('spiral')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.3, 0.025, 0.4])
fig.colorbar(im, cax=cbar_ax)
plt.savefig('compare_streamlines.png')
```
